## Implementación del algoritmo A*

Este notebook muestra una implementación del algoritmo A* en Python.
El algoritmo encuentra el camino más corto en una cuadrícula utilizando
la heurística de distancia Manhattan.


```python
import heapq
import time

# -------------------------------
# Heurística Manhattan
# -------------------------------
def heuristic(a, b):
    return abs(a[0]-b[0]) + abs(a[1]-b[1])


# -------------------------------
# Mostrar el mapa
# -------------------------------
def print_grid(grid, open_nodes, closed_nodes, current):

    rows = len(grid)
    cols = len(grid[0])

    for i in range(rows):

        row = ""

        for j in range(cols):

            pos = (i,j)

            if pos == current:
                row += "A "   # Nodo actual

            elif pos in closed_nodes:
                row += "X "   # Nodo ya explorado

            elif pos in open_nodes:
                row += "O "   # Nodo en Open List

            else:
                row += grid[i][j] + " "

        print(row)

    print("\n")


# -------------------------------
# Algoritmo A* paso a paso
# -------------------------------
def astar_visual(grid, start, goal):

    open_list = [(0,start)]
    open_nodes = {start}

    closed_nodes = set()

    came_from = {}
    g_score = {start:0}

    step = 1

    while open_list:

        current = heapq.heappop(open_list)[1]
        open_nodes.discard(current)

        print(f"Paso {step}")
        print_grid(grid, open_nodes, closed_nodes, current)

        if current == goal:

            path = []

            while current in came_from:
                path.append(current)
                current = came_from[current]

            path.append(start)

            return path[::-1]

        closed_nodes.add(current)

        x,y = current

        neighbors = [(0,1),(0,-1),(1,0),(-1,0)]

        for dx,dy in neighbors:

            nx = x + dx
            ny = y + dy

            if 0 <= nx < len(grid) and 0 <= ny < len(grid[0]):

                if grid[nx][ny] == "#":
                    continue

                neighbor = (nx,ny)

                tentative_g = g_score[current] + 1

                if neighbor not in g_score or tentative_g < g_score[neighbor]:

                    came_from[neighbor] = current
                    g_score[neighbor] = tentative_g

                    f = tentative_g + heuristic(neighbor,goal)

                    heapq.heappush(open_list,(f,neighbor))
                    open_nodes.add(neighbor)

        step += 1

        time.sleep(1)

    return None


# -------------------------------
# MAPA
# -------------------------------
grid = [
["S",".",".",".","."],
[".","#","#","#","."],
[".",".",".","#","."],
["#","#",".",".","."],
[".",".",".","#","G"]
]

start = (0,0)
goal = (4,4)


# -------------------------------
# EJECUTAR ALGORITMO
# -------------------------------
path = astar_visual(grid, start, goal)


# -------------------------------
# Mostrar recorrido final
# -------------------------------
if path:

    print("\n--- META ALCANZADA: RECORRIDO FINAL ---\n")

    for i in range(len(grid)):

        row = ""

        for j in range(len(grid[0])):

            if (i,j) == start:
                row += "S "

            elif (i,j) == goal:
                row += "G "

            elif (i,j) in path:
                row += "* "

            else:
                row += grid[i][j] + " "

        print(row)

    print("\nCoordenadas del camino encontrado:")
    print(path)

else:

    print("No se encontró camino.")
```

## Heurísticas en Inteligencia Artificial

Las heurísticas son funciones utilizadas en inteligencia artificial para estimar qué tan cerca se encuentra un estado de la solución de un problema. En algoritmos de búsqueda informada como A*, las heurísticas permiten reducir el espacio de exploración al priorizar los nodos que parecen más prometedores.  

Una de las heurísticas más utilizadas en problemas de navegación en cuadrículas es la **distancia Manhattan**, la cual calcula la distancia entre dos puntos considerando únicamente movimientos horizontales y verticales.


### Ejemplo 1: Cálculo de la heurística Manhattan



Este ejemplo muestra cómo calcular una estimación de distancia entre dos puntos en una cuadrícula utilizando la heurística Manhattan. Esta heurística se utiliza en algoritmos como A* para estimar qué tan lejos se encuentra el objetivo.


```python
# Ejemplo de heurística Manhattan

def heuristic_manhattan(a, b):
    return abs(a[0] - b[0]) + abs(a[1] - b[1])

start = (13, 27)
goal = (46, 59)

distance = heuristic_manhattan(start, goal)

print("Posición inicial:", start)
print("Objetivo:", goal)
print("Estimación heurística (Manhattan):", distance)
```

### Ejemplo 2: Evaluación de rutas usando la función f(n)


Este ejemplo muestra cómo una inteligencia artificial puede elegir entre varias rutas utilizando la función de evaluación del algoritmo A*. La decisión se basa en el costo real del camino recorrido y una estimación heurística de la distancia restante.


```python
# Simulación de decisión de movimiento: f(n) = g(n) + h(n)

# g: Esfuerzo realizado (terreno difícil aumenta este valor)
# h: Distancia estimada a la base enemiga (Heurística Manhattan)

candidatos_movimiento = {
    "Ruta del Bosque":  {"g": 12, "h": 18}, # Terreno lento pero directo
    "Ruta del Pantano": {"g": 25, "h": 10}, # Muy lento pero ya está cerca
    "Ruta de la Llanura":{"g": 8,  "h": 24}  # Terreno rápido pero se aleja
}

print("--- Evaluación de la Inteligencia de Ruta ---")

for ruta, datos in candidatos_movimiento.items():
    f_total = datos["g"] + datos["h"]
    datos["f"] = f_total
    print(f"{ruta:18} | g={datos['g']:2} + h={datos['h']:2} | Total f(n)={f_total}")

# Selección de la mejor opción
mejor_opcion = min(candidatos_movimiento, key=lambda k: candidatos_movimiento[k]["f"])

print(f"\n🚀 La IA decide avanzar por: {mejor_opcion}")
```

## Teoría de Juegos en Inteligencia Artificial

La teoría de juegos estudia cómo diferentes agentes toman decisiones estratégicas cuando sus resultados dependen también de las decisiones de otros agentes. En inteligencia artificial, este enfoque se utiliza para analizar escenarios donde múltiples agentes interactúan, cooperan o compiten por recursos.

A través de modelos de decisión y matrices de resultados, los algoritmos pueden evaluar distintas estrategias y elegir la opción que maximiza el beneficio esperado. A continuación se presentan algunos ejemplos simples que ilustran cómo la teoría de juegos puede aplicarse en distintos escenarios de sistemas multiagente.


### Ejemplo 1: Robots buscando energía


Este programa simula un escenario de teoría de juegos entre dos robots que deben decidir si buscar energía o esperar. Dependiendo de las decisiones tomadas por cada robot, la energía disponible se distribuye de manera diferente. El programa utiliza un diccionario para representar las posibles combinaciones de decisiones y mostrar cuánta energía obtiene cada robot.


```python
# Simulación de teoría de juegos entre dos robots que buscan energía

energia_total = 10

decisiones = {
    ("Buscar energía", "Buscar energía"): (energia_total/2, energia_total/2),
    ("Buscar energía", "Esperar"): (8, 2),
    ("Esperar", "Buscar energía"): (2, 8),
    ("Esperar", "Esperar"): (0, 0)
}

robot1 = "Buscar energía"
robot2 = "Esperar"

resultado = decisiones[(robot1, robot2)]

print("Robot 1 decide:", robot1)
print("Robot 2 decide:", robot2)

print("\nEnergía obtenida:")
print("Robot 1:", resultado[0])
print("Robot 2:", resultado[1])
```

### Ejemplo 2: Competencia por recursos de red


Este ejemplo simula un escenario de teoría de juegos donde dos servidores compiten por el uso del ancho de banda de una red. Dependiendo de las decisiones tomadas por cada servidor, el recurso disponible se distribuye de manera diferente.

```python

# Ejemplo Competencia por recursos de red
red_game = {

    ("Mucho", "Mucho"): (20, 20),

    # Uno usa mucho ancho de banda y el otro poco
    ("Mucho", "Poco"): (80, 10),
    ("Poco", "Mucho"): (10, 80),

    # Ambos usan poco ancho de banda → red estable
    ("Poco", "Poco"): (50, 50)
}


servidor1 = "Mucho"
servidor2 = "Mucho"

resultado = red_game[(servidor1, servidor2)]

print("Estrategia Servidor 1:", servidor1)
print("Estrategia Servidor 2:", servidor2)

print("\nAncho de banda obtenido:")
print("Servidor 1:", resultado[0], "Mbps")
print("Servidor 2:", resultado[1], "Mbps")

if resultado == (20, 20):
    print("\nEstado de la red: congestión por uso excesivo del ancho de banda.")
```

### Ejemplo 3: Coordinación de rutas entre drones

Este ejemplo simula un escenario de teoría de juegos en el que tres drones deben elegir diferentes rutas de vigilancia. Dependiendo de las decisiones tomadas por cada drone, la cobertura del área puede mejorar o generar redundancia si varios drones seleccionan la misma ruta. El programa muestra cómo la elección de estrategias influye en la eficiencia del sistema de vigilancia.


```python
# Simulación de teoría de juegos entre tres drones que eligen rutas de vigilancia

rutas = {
    ("Ruta A", "Ruta B", "Ruta C"): (6, 6, 6),
    ("Ruta A", "Ruta A", "Ruta B"): (4, 4, 5),
    ("Ruta B", "Ruta C", "Ruta C"): (5, 4, 4),
    ("Ruta A", "Ruta B", "Ruta B"): (6, 3, 3)
}

drone1 = "Ruta A"
drone2 = "Ruta B"
drone3 = "Ruta C"

resultado = rutas[(drone1, drone2, drone3)]

print("Drone 1 elige:", drone1)
print("Drone 2 elige:", drone2)
print("Drone 3 elige:", drone3)

print("\nCobertura obtenida:")
print("Drone 1:", resultado[0])
print("Drone 2:", resultado[1])
print("Drone 3:", resultado[2])
```
